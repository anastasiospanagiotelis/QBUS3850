### QBUS3850 Time Series Analysis 
### The X-13-ARIMA-SEATS decomposition

We will be using the X13 algorithm released by the US Census Bureau.

`x13.x13_arima_analysis(endog[, maxorder, â€¦])`

Perform x13-arima analysis for monthly or quarterly data.

X13 is not part of python. The Python API for X13 is a wrapper for a (Fortran) binary available from the US Census Bureau. Python API reference: https://www.statsmodels.org/stable/generated/statsmodels.tsa.x13.x13_arima_analysis.html


Please [download and install x13 code](https://www.census.gov/data/software/x13as.About_X-13.html) from the US Census Bureau and confirm the following test case works.

Linux (Debian): apt-get install x13as

Windows: follow instructions [here](https://www.census.gov/data/software/x13as.Win_X-13.html) 



```python
import pandas as pd
import numpy as np
from statsmodels.tsa import x13

np.random.seed( 0x123abc )
values = np.random.rand(48)
indices = pd.date_range( start='2020-01-01', periods=48, freq='M' )
ts = pd.DataFrame( data=values, index=indices )

## Should not throw any errors.
model = x13.x13_arima_analysis( ts )

model.trend.plot()
#print( model.results )
```
